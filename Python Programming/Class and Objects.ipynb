```python
class Student:
  def __init__(self, name, roll_no, marks):
    self.name = name;
    self.roll_no = roll_no
    self.marks = marks

  def calculate_total_marks(self):
    return sum(self.marks)

  def calculate_percentage(self):
    obtained_marks = self.calculate_total_marks()
    total_marks = len(self.marks) * 100
    return (obtained_marks / total_marks ) * 100

  def __str__(self):
    return f"Name= {self.name}, Roll No:{self.roll_no}, Marks={self.marks}"

```

```python
stud1 = Student("Ram",1001,[90,92,93,94])
stud2 = Student("Krishna",1001,[90,97,91,54])
stud3 = Student("Balaram",1001,[90,65,92,94])


students = [stud1, stud2, stud3]
D = {}

for student in students:
  D[student.name] = student.calculate_percentage()

print(D)

max_score = max(D.values())
print(max_score)
print([name for name, score in D.items() if score == max_score][0])
```

# Attributes / Properties / Variables

There are two types of variables

1. Instance Variables
2. Static Variables

## Instance Variables

> Instance Variables are the variables which are unique for every object we creates

```python
class A:
  def __init__(self, x):
    self.x = x

obj_01 = A(10)
obj_01.x = 100
obj_02 = A(20)
print(obj_02.x)
```

## Static Variables

> Static Variables are the variable which are shared across multiple instances

```python
class A:
  company = "Infyni"
  def __init__(self, x):
    self.x = x


  @staticmethod
  def get_company():
    return A.company;

obj_01 = A(10)
obj_02 = A(20)
print(A.get_company())
```
