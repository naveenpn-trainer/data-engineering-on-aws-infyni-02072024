## Default Arguments

```python
def calculate_percentage(obtained_marks, total_marks=100):
  percentage = (obtained_marks/total_marks) * 100
  return f"{percentage:.2f}%"

```

```python
print(calculate_percentage(35))
print(calculate_percentage(45))
print(calculate_percentage(55,200))
```

# Lambda Functions

> A lambda function is a small anonymous function defined using lambda keyword

*Syntax*

`
lambda arguments, expression
`

```python
def increment_by_10(number):
  return number + 10
```

```python
increment_by_10_lambda = lambda number:number+10
increment_by_10_lambda(30)
```

```python
add_given_numbers = lambda x,y : x+y
add_given_numbers(9,1)
```

```python
L = [1,2,3,4,5,6]
new_list = []
for e in L:
  new_list.append(e*3)
print(new_list)
```

```python
def mul_by_3(L):
  new_list = []
  for e in L:
    new_list.append(e*3)
  return new_list


L = [1,2,3,4,5,6]
print(mul_by_3(L))
```

```python
print(list(map(lambda e:e*3, L)))
print(list(filter(lambda e:e%2==0, L)))
```

```python
divisible_by_2 = lambda e:e%2==0
greater_than_3 = lambda e:e>3

print(list(filter(divisible_by_2, L)))
```
