**Problem Statement**

Give a list of numbers

`
number = int(input("Enter a number"))
L = [1,3,6,7,3,4,6,3]
`

Write a program to find how many times a given number exist

```python
L=[1,3,2,4,6,3,5]
number = int(input("Enter a number"));
print(L.count(number))
count=0;
for e in L:
    if e==number:
      count+=1
print(count)
```

# Nested list

```python
L = [1,2,3,[4,5,6],[5,6],10]
print(str(type(L)) == "<class 'list'>")
print(type(L[3]))
for e in L:
  if type(e) == str(list):
    print(e)
```

# Dictionary

## Create Dictionary

1. {"key":"value"} - Recommended way
2. dict(**kwarg)
3. dict(mapping, **kwarg)
4. dict(iterable)
5. dict(iterable, **kwarg)

```python
D1 = {"apple":"red", "grapes":"green"}
print(type(D1))

D2 = dict(apple="red",grapes="green")
print(type(D2))


D3 = dict(D1,oranges="orange")
print(type(D3))


D4 = dict([("apples","red"),("grapes","green")])
print(type(D4))

D5 = dict([("apples","red"),("grapes","green")],oranges="orange")
print(type(D5))
```

## Iterating Dictionary

```python
D = {"k1":"v1","k2":"v2","k3":"v3"}
D
```

### Iterating through the keys

```python
for k in D:
  print(D[k])
```

```python
for k in D.keys():
  print(k)
```

### Iterating thorough the values

```python
for v in D.values():
  print(v)
```

### Iterating through the key and values

```python
for k,v in D.items():
  print(k,v)
```
